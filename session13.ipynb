```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_classification
```

```python
# Step 1: Generate synthetic data
# 100 samples, 2 informative features for clear 2D plot, 2 classes
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
```

```python
# Step 2: Visualize the data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=30)
plt.title("Email Dataset: Spam vs Not Spam")
plt.xlabel("Spam")
plt.ylabel("Not Spam")
plt.show()
```

```python
# Step 3: Train SVM Model
# Create a linear SVM model
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)
```

```python
# Step 4: Visualize the Decision Boundary and Support Vectors
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=30)
plt.title("SVM Decision Boundary for Spam Classification")
```

```python
# Retrieve the model's weights and intercept
w = model.coef_[0]
b = model.intercept_[0]
```

```python
# Create a range of x values to plot the decision boundary
x_plot = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
# Calculate the corresponding y values for the decision boundary
y_plot = -(w[0] / w[1]) * x_plot - b / w[1]
```

```python
# Plot the decision boundary
plt.plot(x_plot, y_plot, 'k-', label="Decision Boundary")
```

```python

# Plot margins
margin = 1 / np.sqrt(np.sum(w ** 2))
y_margin_positive = y_plot + margin
y_margin_negative = y_plot - margin
plt.plot(x_plot, y_margin_positive, 'k--', label="Margin")
plt.plot(x_plot, y_margin_negative, 'k--')
```

```python
# Highlight the support vectors
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='k', label="Support Vectors")

plt.xlabel("Spam")
plt.ylabel("Not Spam")
plt.legend()
plt.show()
```
