```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
# 1. Create dataset
data = {
    'Age': [25, 30, 22, 28, 35, 31, 26, 21, 29, 32],
    'Batting Average': [40, 50, 30, 35, 55, 45, 42, 33, 48, 41],
    'Wickets Taken': [5, 10, 0, 8, 12, 7, 6, 0, 11, 9],
    'Strike Rate': [120, 110, 130, 125, 100, 105, 115, 135, 108, 118],
    'Is Captain': [1, 0, 0, 1, 0, 1, 0, 0, 0, 0],
    'Selected': [1, 1, 0, 1, 1, 1, 1, 0, 1, 1]
}
```

```python
# 2. Create DataFrame
df = pd.DataFrame(data)
```

```python
# Features and Target
X = df[['Age', 'Batting Average', 'Wickets Taken', 'Strike Rate', 'Is Captain']]
y = df['Selected']
```

```python
# 3. Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 4. Create & Train Decision Tree
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
# 5. Predictions & Accuracy
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# 6. Plot Decision Tree
plt.figure(figsize=(12,8))
tree.plot_tree(
    clf,
    filled=True,
    feature_names=['Age', 'Batting Average', 'Wickets Taken', 'Strike Rate', 'Is Captain'],
    class_names=['Not Selected', 'Selected'],
    rounded=True
)
plt.show()
```
